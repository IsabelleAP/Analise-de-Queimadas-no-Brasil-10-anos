```python
#importação das bibliotecas
import pandas as pd
```

```python
#função para limpar os dados de cada arquivo csv anual e salvar como novo arquivo

def limpar_dados_incendio(nome_arquivo, novo_arquivo):
    df = pd.read_csv(nome_arquivo)
    df = df.drop(columns=["id_bdq", "foco_id", "lat", "lon"])
    df[["data", "hora"]] = df["data_pas"].str.split(" ", expand=True)
    df = df.drop(columns=["data_pas", "hora"])
    df[["ano", "mês", "dia"]] = df["data"].str.split("-", expand=True)
    df = df.drop(columns=["data", "pais"])
    df = df.drop_duplicates()

    df.to_csv(novo_arquivo, index=False)

    return df

df = limpar_dados_incendio("focos_br_ref_2014.csv", "2014_limpo.csv")
display(df)
```

```python
#função para limpar os dados de cada arquivo csv mensal e salvar como novo arquivo

def limpar_dados_2024(nome_arquivo, novo_arquivo):
    dados = pd.read_csv(nome_arquivo)
    dados = dados.drop(columns=["id", "satelite", "municipio_id", "estado_id", 
                                "lat", "lon", "pais_id", "pais", "numero_dias_sem_chuva", 
                                "precipitacao", "risco_fogo", "frp"])
    dados[["data", "gmt"]] = dados["data_hora_gmt"].str.split(" ", expand=True)
    dados[["ano","mês","dia"]] = dados["data"].str.split("-", expand=True)
    dados = dados.drop(columns=["data_hora_gmt","data","gmt"])
    dados[['municipio', 'estado']] = dados[['estado', 'municipio']]
    dados = dados.rename({'estado':'municipio', 'municipio':'estado'}, axis=1)
    dados = dados.drop_duplicates()
    
    dados.to_csv(novo_arquivo, index=False)

    return dados

dados = limpar_dados_2024("focos_mensal_br_202410.csv", "202410_limpo.csv")
display(dados)
```
